```python
import yfinance as yf
import datetime

# Get the stock quote using yfinance
df = yf.download('AAPL', start='2012-01-01', end=datetime.datetime.now()) # Changed to import datetime and then use datetime.datetime.now()

# Show the data
df
```

```python
plt.figure(figsize=(16,6))
plt.title('Close Price History')
plt.plot(df['Close'])
plt.xlabel('Date', fontsize=18)
plt.ylabel('Close Price USD ($)', fontsize=18)
plt.show()
```

```python
# Create a new dataframe with only the 'Close' column
data = df[['Close']] # Changed to use double brackets to select the column
# Convert the dataframe to a numpy array
dataset = data.values
# Get the number of rows to train the model on
training_data_len = int(np.ceil( len(dataset) * .95 ))

training_data_len
```

```python
# Scale the data
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(dataset)

scaled_data
```

```python
# Create the training data set
# Create the scaled training data set
train_data = scaled_data[0:int(training_data_len), :]
# Split the data into x_train and y_train data sets
x_train = []
y_train = []

for i in range(60, len(train_data)):
    x_train.append(train_data[i-60:i, 0])
    y_train.append(train_data[i, 0])
    if i<= 61:
        print(x_train)
        print(y_train)
        print()

# Convert the x_train and y_train to numpy arrays
x_train, y_train = np.array(x_train), np.array(y_train)

# Reshape the data
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
# x_train.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM

# Build the LSTM model
model = Sequential()
model.add(LSTM(128, return_sequences=True, input_shape= (x_train.shape[1], 1)))
model.add(LSTM(64, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(x_train, y_train, batch_size=1, epochs=2)
```

```python
# Create the testing data set
# Create a new array containing scaled values from index 1543 to 2002
test_data = scaled_data[training_data_len - 60: , :]
# Create the data sets x_test and y_test
x_test = []
y_test = dataset[training_data_len:, :]
for i in range(60, len(test_data)):
    x_test.append(test_data[i-60:i, 0])

# Convert the data to a numpy array
x_test = np.array(x_test)

# Reshape the data
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1 ))

# Get the models predicted price values
predictions = model.predict(x_test)
predictions = scaler.inverse_transform(predictions)

# Get the root mean squared error (RMSE)
rmse = np.sqrt(np.mean(((predictions - y_test) ** 2)))
rmse
```

```python
# Plot the data
train = data[:training_data_len]
valid = data[training_data_len:]
valid['Predictions'] = predictions
# Visualize the data
plt.figure(figsize=(16,6))
plt.title('Model')
plt.xlabel('Date', fontsize=18)
plt.ylabel('Close Price USD ($)', fontsize=18)
plt.plot(train['Close'])
plt.plot(valid[['Close', 'Predictions']])
plt.legend(['Train', 'Val', 'Predictions'], loc='lower right')
plt.show()
```

```python
from datetime import datetime, timedelta
# Function to predict stock prices for a specified period and return predictions
def predict_stock_prices(ticker, days_to_predict):
    # Get the stock quote using yfinance
    df = yf.download(ticker, start='2012-01-01', end=datetime.now())
    data = df[['Close']]
    dataset = data.values

    training_data_len = int(np.ceil(len(dataset) * .95))

    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(dataset)

    # Create the testing data set for prediction
    test_data = scaled_data[training_data_len - 60:, :]
    x_test = []
    y_test = dataset[training_data_len:, :]  # Actual values for comparison

    for i in range(60, len(test_data)):
        x_test.append(test_data[i - 60:i, 0])

    x_test = np.array(x_test)
    x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))

    predictions = model.predict(x_test)
    predictions = scaler.inverse_transform(predictions)

    # Extend predictions to the specified number of days
    last_60_days = scaled_data[-60:]  # Use the last 60 days as input
    future_predictions = []

    for i in range(days_to_predict):
        x = np.array([last_60_days])
        x = np.reshape(x, (x.shape[0], x.shape[1], 1))
        predicted_price = model.predict(x)
        future_predictions.append(scaler.inverse_transform(predicted_price)[0,0])
        last_60_days = np.append(last_60_days[1:], predicted_price, axis=0) # slide window

    future_predictions = np.array(future_predictions).reshape(-1,1)

    # Plot the graph with extended predictions
    train = data[:training_data_len]
    valid = data[training_data_len:]
    valid['Predictions'] = predictions

    # Date range for the extended predictions
    last_date = df.index[-1]
    future_dates = [last_date + timedelta(days=i) for i in range(1, days_to_predict + 1)]

    plt.figure(figsize=(16, 6))
    plt.title(f'Price Prediction for {ticker} ({days_to_predict} days)')
    plt.xlabel('Date', fontsize=18)
    plt.ylabel('Close Price USD ($)', fontsize=18)
    plt.plot(train['Close'], label='Train')
    plt.plot(valid[['Close', 'Predictions']])
    plt.plot(future_dates, future_predictions, label=f'{days_to_predict} Days Forecast')
    plt.legend(['Train', 'Val', 'Predictions', f'{days_to_predict} Days Forecast'], loc='lower right')
    plt.show()

    # Print the forecasted prices
    print(f'Predicted Prices for {days_to_predict} days: {future_predictions.flatten()}')

    return future_predictions

# Example usage: Predict AAPL stock prices for the next 7, 15, and 30 days
predicted_prices_7d = predict_stock_prices('AAPL', 7)
predicted_prices_15d = predict_stock_prices('AAPL', 15)
predicted_prices_30d = predict_stock_prices('AAPL', 30)
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from textblob import TextBlob
from datetime import timedelta

# Download all relevant stock data for Apple (AAPL)
def download_stock_data():
    # Adjust date range to be broader to account for potential gaps in news dates
    stock_data = yf.download('AAPL', start='2017-01-01', end=pd.Timestamp.now().strftime('%Y-%m-%d'))
    stock_data.dropna(inplace=True)
    stock_data.index = pd.to_datetime(stock_data.index)
    print(stock_data.head())  # Debugging output to check stock data
    return stock_data


# Function to analyze the news impact on stock prices by comparing with the closest available stock date
def analyze_news_impact(headline, date, stock_data):
    # Convert the news date to pandas Timestamp
    news_date = pd.to_datetime(date)
    stock_dates = stock_data.index

    # Step 1: Check if the news date exists in the stock data
    if news_date in stock_dates:
        closest_date = news_date
    else:
        # Step 2: If the exact news date doesn't exist, find the closest available date after the news date
        future_dates = stock_dates[stock_dates > news_date]
        if not future_dates.empty:
            closest_date = future_dates[0]
        else:
            print(f"No stock data found after {news_date}. Skipping analysis for this news.")
            return

    # Get stock prices for the news date (if available)
    if news_date in stock_dates:
        # Access the first element of the Series to get the numeric value
        price_on_news_date = stock_data.loc[news_date, 'Close']

        # Check if price_on_news_date is a Series and get the first element if it is
        if isinstance(price_on_news_date, pd.Series):
            price_on_news_date = price_on_news_date.iloc[0]

        print(f"Stock Close Price on News Date ({news_date}): {price_on_news_date:.2f} USD")
    else:
        print(f"No stock data available on {news_date}. Using closest available stock date instead.")


    # Get stock prices for the next 7 consecutive days after the closest date
    stock_slice = stock_data.loc[closest_date:closest_date + timedelta(days=7)]

    # Print next 7 entries after the news date
    next_3_entries = stock_slice.head(7)
    print(f"Next 7 Stock Entries After News Date ({news_date}):")
    print(next_3_entries[['Close']])

    # Analyze the sentiment of the news
    sentiment = TextBlob(headline).sentiment.polarity

    # Output the results
    print(f"Headline: {headline}")
    print(f"News Date: {date}")
    print(f"Closest Stock Date: {closest_date}")

    print(f"News Sentiment: {'Positive' if sentiment > 0 else 'Negative' if sentiment < 0 else 'Neutral'}")
    print("=" * 50)

    # Plot the stock price movement if there's enough data
    if len(stock_slice) >= 7:
        plt.figure(figsize=(10, 5))
        plt.plot(stock_slice.index, stock_slice['Close'], label='Stock Price')
        plt.axvline(x=closest_date, color='red', linestyle='--', label='News Date')
        plt.title(f"Stock Price Movement for AAPL (7 days after news)")
        plt.xlabel('Date')
        plt.ylabel('Close Price USD ($)')
        plt.legend()
        plt.show()


# Download stock data for AAPL
apple_stock_data = download_stock_data()

# Randomly select 5 Apple-related news articles from the dataset
apple_related_news = df[df['Headlines'].str.contains("Apple", case=False)].sample(5)

# Analyze each randomly selected news article
for idx, row in apple_related_news.iterrows():
    analyze_news_impact(row['Headlines'], row['Time'], apple_stock_data)
```

## Stock Ticker Extraction Methodology:

The code uses the `yfinance` library to download historical stock data. The ticker symbol ('AAPL' for Apple) is explicitly hardcoded in the `download_stock_data` and `analyze_news_impact` functions. While straightforward, a more robust approach might involve a separate function to extract tickers from the news headlines based on regular expressions or named entity recognition techniques to automatically identify companies and their corresponding tickers.


## Financial Insights and Forecasting Results:

The notebook employs an LSTM model to predict future stock prices. The model is trained on historical stock data (`Close` prices), scaled using `MinMaxScaler`, and then used to predict future prices.

**Model Evaluation:**
* The code calculates the Root Mean Squared Error (RMSE). This metric helps evaluate the accuracy of the predictions.

* **Limitations:** The LSTM model's performance depends heavily on the quality and quantity of the input data. The evaluation metrics, like RMSE, indicate how well the model fits the historical data, but provide no guarantee of future performance.  External factors, which are not considered by the model, can significantly affect actual stock prices.

* **Forecasting:** The `predict_stock_prices` function extends predictions for the next 7, 15, and 30 days using the last 60 days as input. A rolling window approach is used to get continuous forecasts, which is a reasonable approach.  It is important to remember that these are probabilistic forecasts and, thus, not guarantees.


## Correlation of News with Stock Performance:

The code attempts to correlate news headlines with stock performance:

1. **News Sentiment Analysis:** It uses `TextBlob` for sentiment analysis of the headlines. This helps determine whether the news is predominantly positive, negative, or neutral.

2. **Date Alignment:**  A crucial step is aligning the news date with the closest available stock data date, which is a common challenge since news may not always align precisely with trading days. It tries to handle the case where the exact news date isn't in the stock data by using the nearest subsequent date.

3. **Impact Analysis:** The code compares the stock price change over 15 days (or 7 days) after the news date with the sentiment score. A potential refinement would involve calculating the average price change over a longer timeframe.

4. **Visualization:** Plots show the stock price movement around the news date, aiding visual inspection.

**Limitations:**
* **Causality:** Correlation doesn't equal causation. Just because a positive news headline is released and the stock price rises over 15 days afterward, does not mean that the news directly caused the increase. Many external factors and market forces can affect a stock price at any point in time.
* **Sentiment Accuracy:**  The `TextBlob` sentiment analysis may not accurately capture the nuances of financial news headlines. More sophisticated sentiment analysis tools or models would be more useful to consider.
* **Short-Term Analysis:** The analysis looks at relatively short periods (7, 15, 30 days). Market reactions to news can occur over more extended durations, sometimes much longer.
